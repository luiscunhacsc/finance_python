```python
import pandas as pd
import matplotlib.pyplot as plt

# URL to the new extended CSV file hosted on GitHub
url = 'https://raw.githubusercontent.com/luiscunhacsc/finance_python/main/financial_data_ratios_intro__extended.csv'

# Reading data into a pandas DataFrame
data = pd.read_csv(url)

# Converting the 'Date' column to datetime format
data['Date'] = pd.to_datetime(data['Date'])

# Calculating Profitability Ratios
data['Net Profit Margin'] = (data['Net Income'] / data['Revenue']) * 100

# Plotting Net Profit Margin over time
plt.figure(figsize=(10, 5))
plt.plot(data['Date'], data['Net Profit Margin'], marker='o', linestyle='-', color='b', label='Net Profit Margin')
plt.xlabel('Date')
plt.ylabel('Net Profit Margin (%)')
plt.title('Net Profit Margin Over Time')
plt.legend()
plt.grid(True)
plt.show()
```
