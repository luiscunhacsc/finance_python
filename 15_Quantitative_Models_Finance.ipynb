**Applications:** Portfolio optimization, risk management, option pricing, and more.


# Probability Distributions in Finance

**Definition:** Probability distributions describe how the values of a random variable are distributed.

**Types:**
- **Normal Distribution:** Symmetrical, bell-shaped curve.
- **Lognormal Distribution:** Distribution of a variable whose logarithm is normally distributed.
- **Poisson Distribution:** Models the probability of a given number of events occurring in a fixed interval of time.

**Applications:** Modeling asset returns, risk assessment, and financial forecasting.


# Normal Distribution

**Characteristics:** Symmetrical, bell-shaped curve with mean (\(\mu\)) and standard deviation (\(\sigma\)).

**Formula:**

$$
f(x \mid \mu, \sigma) = \frac{1}{\sigma \sqrt{2\pi}} \exp\left(-\frac{(x - \mu)^2}{2\sigma^2}\right)
$$





```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Parameters
mu = 0
sigma = 1

# Generate data points
x = np.linspace(mu - 3*sigma, mu + 3*sigma, 100)
y = norm.pdf(x, mu, sigma)

# Plot the distribution
plt.figure(figsize=(14, 7))
plt.plot(x, y, label='Normal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Normal Distribution (μ=0, σ=1)')
plt.legend()
plt.grid(True)
plt.show()
```

# Lognormal Distribution

**Characteristics:** Distribution of a variable whose logarithm is normally distributed.

**Formula:**

$$
f(x \mid \mu, \sigma) = \frac{1}{x\sigma\sqrt{2\pi}} \exp\left(-\frac{(\ln(x) - \mu)^2}{2\sigma^2}\right)
$$



**Applications:** Modeling stock prices, option pricing, and risk management.


```python
import numpy as np
import matplotlib.pyplot as plt

# Parameters
mu = 0
sigma = 0.25

# Generate data points, avoiding x=0
x = np.linspace(0.001, 3, 100)  # Start x from 0.001 instead of 0
y = (1 / (x * sigma * np.sqrt(2 * np.pi))) * np.exp(- (np.log(x) - mu) ** 2 / (2 * sigma ** 2))

# Plot the distribution
plt.figure(figsize=(10, 5))
plt.plot(x, y, label='Lognormal Distribution')
plt.xlabel('Value')
plt.ylabel('Probability Density')
plt.title('Lognormal Distribution (μ=0, σ=0.25)')
plt.legend()
plt.grid(True)
plt.show()

```

# Poisson Distribution

**Characteristics:** Models the probability of a given number of events occurring in a fixed interval of time or space.

**Formula:**

$$
P(X = k) = \frac{\lambda^k e^{-\lambda}}{k!}
$$

**Applications:** Modeling the number of trades per day, number of defaults, and other discrete events.


```python
from scipy.stats import poisson

# Parameter
lambda_ = 5

# Generate data points
x = np.arange(0, 15)
y = poisson.pmf(x, lambda_)

# Plot the distribution
plt.figure(figsize=(10, 5))
plt.bar(x, y, label='Poisson Distribution')
plt.xlabel('Number of Events')
plt.ylabel('Probability')
plt.title('Poisson Distribution (λ=5)')
plt.legend()
plt.grid(True)
plt.show()
```

# Basic Statistics in Finance

- **Mean:** Average value of a dataset.
- **Median:** Middle value of a dataset.
- **Variance:** Measure of the dispersion of a dataset.
- **Standard Deviation:** Square root of the variance, indicating the spread of data points.


```python
# Generate a dataset of stock returns
returns = np.random.normal(0.01, 0.02, 1000)

# Calculate basic statistics
mean_return = np.mean(returns)
median_return = np.median(returns)
variance_return = np.var(returns)
std_deviation_return = np.std(returns)

print(f"Mean Return: {mean_return:.4f}")
print(f"Median Return: {median_return:.4f}")
print(f"Variance of Return: {variance_return:.6f}")
print(f"Standard Deviation of Return: {std_deviation_return:.4f}")

# Visualization
plt.figure(figsize=(14, 7))
plt.hist(returns, bins=50, alpha=0.75, label='Stock Returns')
plt.axvline(mean_return, color='r', linestyle='dashed', linewidth=2, label='Mean')
plt.axvline(median_return, color='g', linestyle='dashed', linewidth=2, label='Median')
plt.xlabel('Return')
plt.ylabel('Frequency')
plt.title('Histogram of Stock Returns')
plt.legend()
plt.grid(True)
plt.show()
```
